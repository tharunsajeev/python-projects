# ATTENDANCE MARKER

```python
import datetime
from collections import defaultdict

class Student:
    def __init__(self, roll_no, name):
        self.roll_no = roll_no
        self.name = name
        self.attendance = []
        self.present = 0
        self.absent = 0

    def mark_attendance(self, date, status):
        status = status.upper()
        if status not in ["P", "A"]:
            print("Invalid status! Use 'P' for Present and 'A' for Absent")
            return
        self.attendance.append((date, status))
        print(f"Attendance marked for {self.name} on {date} as {status}")

    def view_attendance(self):
        print(f"\nAttendance for {self.name} ({self.roll_no}):")
        for date, status in self.attendance:
            print(f"  {date} : {status}")

    def attendance_summary(self):
        self.present = 0
        self.absent = 0
        for date, status in self.attendance:
            if status == "P":
                self.present += 1
            elif status == "A":
                self.absent += 1
        print(f"\nAttendance status of {self.name}: Present={self.present}, Absent={self.absent}")

    def calculate_percentage(self):
        total = self.present + self.absent
        if total == 0:
            return 0
        return (self.present / total) * 100

    def weekly_report(self):
        print(f"\nWeekly Report for {self.name}:")
        weekly_data = defaultdict(list)

        for date, status in self.attendance:
            week_number = datetime.datetime.strptime(date, "%Y-%m-%d").isocalendar()[1]
            weekly_data[week_number].append(status)

        for week, records in weekly_data.items():
            present = records.count("P")
            absent = records.count("A")
            total = len(records)
            percentage = (present / total) * 100
            print(f"  Week {week}: Present={present}, Absent={absent}, Attendance={percentage:.2f}%")


class AttendanceReport:
    def __init__(self, student: Student):
        student.attendance_summary()
        percentage = student.calculate_percentage()

        if percentage == 0:
            print("The student has not attended any class")
        elif percentage >= 75:
            print("The student has above 75% attendance")
        else:
            print("The student has below 75% attendance")


# ---------------- Default students ----------------
students = {
    "101": Student("101", "Tharun"),
    "102": Student("102", "Aadith"),
    "103": Student("103", "Harsh")
}

# Adding some default attendance records
students["101"].attendance = [("2025-08-22", "P"), ("2025-08-23", "A"), ("2025-08-24", "P")]
students["102"].attendance = [("2025-08-22", "A"), ("2025-08-23", "A"), ("2025-08-24", "P")]
students["103"].attendance = [("2025-08-22", "P"), ("2025-08-23", "P"), ("2025-08-24", "P")]

# ---------------- Teacher-side menu ----------------
while True:
    print("\n===== Attendance System =====")
    print("1. Add New Student")
    print("2. Mark Attendance")
    print("3. View Attendance")
    print("4. Attendance Summary & Report")
    print("5. Weekly Report")
    print("6. Exit")
    print("7. View All Students’ Summary")
    print("8. View Student List")   # NEW OPTION
    choice = input("Enter your choice: ")

    if choice == "1":
        roll_no = input("Enter new student roll number: ")
        name = input("Enter student name: ")
        if roll_no not in students:
            students[roll_no] = Student(roll_no, name)
            print(f"Student {name} (Roll No: {roll_no}) added successfully.")
        else:
            print("Student already exists!")

    elif choice == "2":
        roll_no = input("Enter roll number: ")
        if roll_no in students:
            date = input("Enter date (YYYY-MM-DD) or press Enter for today: ")
            if date.strip() == "":
                date = datetime.date.today().strftime("%Y-%m-%d")
            status = input("Enter status (P/A): ")
            students[roll_no].mark_attendance(date, status)
        else:
            print("Student not found!")

    elif choice == "3":
        roll_no = input("Enter roll number: ")
        if roll_no in students:
            students[roll_no].view_attendance()
        else:
            print("Student not found!")

    elif choice == "4":
        roll_no = input("Enter roll number: ")
        if roll_no in students:
            report = AttendanceReport(students[roll_no])
        else:
            print("Student not found!")

    elif choice == "5":
        roll_no = input("Enter roll number: ")
        if roll_no in students:
            students[roll_no].weekly_report()
        else:
            print("Student not found!")

    elif choice == "6":
        print("Exiting...")
        break

    elif choice == "7":
        print("\n===== All Students’ Attendance Summary =====")
        for stu in students.values():
            stu.attendance_summary()
            percentage = stu.calculate_percentage()
            print(f"  -> Attendance %: {percentage:.2f}%")

    elif choice == "8":
        print("\n===== Student List =====")
        print(f"Total Students: {len(students)}")
        for roll_no, stu in students.items():
            print(f"  Roll No: {roll_no}, Name: {stu.name}")

    else:
        print("Invalid choice. Try again.")
```
